```python
%matplotlib inline
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = (20, 15)
pylab.rcParams['font.size'] = 24
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

labels = ["SAPS-I", "SOFA","In Hospital Length of Stay", "Survival Rate After Discharge"]
ylabels = ["SAPS-I", "SOFA", "Mean values", "Average days after discharge"]

#set k = 41 for length of stay and k=42 for survival rate
k = 42
patient_dataset = pd.read_csv('patients_data.csv')
p = patient_dataset.iloc[:, k]
p[np.isnan(p)] = 0
mp = np.mean(p)
sp = np.std(p)

subtype0_dataset = pd.read_csv('pdata05.csv')
s0 = subtype0_dataset.iloc[:, k].values
s0[np.isnan(s0)] = 0

subtype1_dataset = pd.read_csv('pdata15.csv')
s1 = subtype1_dataset.iloc[:, k].values
s1[np.isnan(s1)] = 0

subtype2_dataset = pd.read_csv('pdata25.csv')
s2 = subtype2_dataset.iloc[:, k].values
s2[np.isnan(s2)] = 0

subtype3_dataset = pd.read_csv('pdata35.csv')
s3 = subtype3_dataset.iloc[:, k].values
s3[np.isnan(s3)] = 0

subtype4_dataset = pd.read_csv('pdata45.csv')
s4 = subtype4_dataset.iloc[:, k].values
s4[np.isnan(s4)] = 0


sm0 = np.mean(s0)
sm1 = np.mean(s1)
sm2 = np.mean(s2)
sm3 = np.mean(s3)
sm4 = np.mean(s4)

if k == 39:
    label1 = labels[0]
    label2 = ylabels[0]
elif k == 40:
    label1 = labels[1]
    label2 = ylabels[1]
elif k == 41:
    label1 = labels[2]
    label2 = ylabels[2]
elif k == 42:
    label1 = labels[3]
    label2 = ylabels[3]

f = 30

plt.title('Average ' + label1, fontsize = f)
plt.bar(["Subtype 0","Subtype 1","Subtype 2","Subtype 3","Subtype 4"],
        [sm0,sm1,sm2,sm3,sm4], 
        color = ['cornflowerblue', 'sandybrown', 'mediumspringgreen','red', 'mediumorchid'],
        edgecolor = 'black',
        alpha = 0.7, width = 0.5)

plt.legend(loc='best')
plt.ylabel(label2, fontsize = f)
plt.xlabel("Subtypes", fontsize = f)
plt.xticks(fontsize = f)
plt.grid(True)

plt.show()

print("Subtype's " + label1 + " mean: ", [sm0,sm1,sm2,sm3,sm4])
print("Population " + label1 + " mean: ", mp)

```

```python
patient_dataset = pd.read_csv('patients_data.csv')
P = patient_dataset.iloc[:, 42]
P[np.isnan(P)] =''
P = np.array(P)
dead = []
for i in range(len(P)):
    if P[i] != '':
        dead.append(P[i])
        
deadnew = []
for j in range(len(dead)):
    if dead[j] < 365:
        deadnew.append(dead[j])

print(len(dead))
print(len(deadnew))
```
